```python
def files_from_folder( folder ):
    files = list( map( str, pathlib.Path( folder ).glob('*.tif')) )
    return files
```

```python
def rasters_from_files( files ):
    rasters = []
    for file in files:
        
        with gw.open(file) as src:
            rasters.append( src )
    return rasters
    
```
